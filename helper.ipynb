```python
#import libraries
import pandas as pd
import numpy as np
import streamlit as st
```

```python
#read in data
df=pd.read_csv('./gcar_data.csv',index_col=0)
```

```python
non_num= df[~df['year'].apply(lambda x: x.isnumeric())]
non_num_count=len(non_num)
print(non_num_count)
```

```python
df_cleaned = df.dropna()
df_cleaned.isnull().sum()
```

```python
df_cleaned.shape
```

```python
df_cleaned.to_csv('./dataset/cleaned_data.csv', index=False)
```

```python
df1 = df_cleaned
```

```python
non_num = df_cleaned[~df_cleaned['year'].astype(str).str.isnumeric()]
non_num_count=len(non_num)
print(non_num_count)
```

```python
df1['year'] = pd.to_numeric(df1['year'], errors='coerce')
df1 = df1.dropna(subset=['year'])
df1.shape
```

```python
df1.head()
```

```python
df1['Car_Number'] = range(1, len(df1) + 1)
df1 = df1[['Car_Number'] + [col for col in df_cleaned.columns if col != 'Car_Number']]
df1 = df1.loc[:, ~df1.columns.duplicated()]
```
